```python
import pandas as pd
import numpy as np
messages=pd.read_csv('SpamCollection.csv',names=["label","message"])
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]

for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=" ".join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text  import CountVectorizer
```

```python
## for binary BOW, we are enabling the binary parameter
cv=CountVectorizer(max_features=100,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
## for binary BOW, we are enabling the binary parameter, and with ngram
cv=CountVectorizer(max_features=100,binary=True,ngram_range=(1,1))
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```
